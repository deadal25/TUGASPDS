```python
#Latihan (13)
#Hitung korelasi dataset

df.corr(numeric_only=True)
```

```python
#latihan (14)
#Simpulan Sementara Hasil Korelasi di latihan (13)

# Hasil korelasi berupa matriks korelasi yang menunjukkan sejauh mana hubungan antara berbagai pasangan kolom numerik dalam dataset Anda.
# Nilai koefisien korelasi dalam matriks ini berkisar antara -1 hingga 1. Nilai positif mendekati 1 menunjukkan hubungan positif, nilai negatif mendekati -1 menunjukkan hubungan negatif, dan nilai mendekati 0 menunjukkan hubungan lemah atau tidak ada hubungan linear antara variabel-variabel tersebut.
# Intinya kesimpulannya itu Semakin tinggi nilai absolut dari koefisien korelasi, semakin kuat hubungan linear antara dua variabel tersebut.
```

```python
#Latihan (15)
#Hitung korelasi dataset untuk kolom PetalLengthCm, PetalWidthCm

korelasi_duakolom = df['PetalLengthCm'].corr(df['PetalWidthCm'])
print(f"Korelasi antara PetalLengthCm dan PetalWidthCm: {korelasi_duakolom}")
```
